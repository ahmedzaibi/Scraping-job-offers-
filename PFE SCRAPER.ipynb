##  KEEJOB SCRAPER

```python
import pandas as pd
from urllib.parse import unquote
import re
from selenium import webdriver
from selenium.webdriver.common import keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import csv
csv_columns = ['Publiée le', 'Type de poste','Expérience', 'Étude', 'Rémunération proposée', 'Disponibilité', 'Langues', 'Mobilité']
browser=webdriver.Chrome("C:\selinium\chromedriver.exe")
wait = WebDriverWait(browser, 10)
elementsall=[]
keys=[]
for page in range (1,100):
    browser.get("https://www.keejob.com/offres-emploi/?page="+str(page))
    jobs1=browser.find_elements(By.XPATH,'//div//h6//a')
    jobs=[]
    for job in jobs1:
        jobs.append(job.get_attribute('href'))
    print(len(jobs))
    for i in range(len(jobs)):
        print(i+1)
        browser.get(jobs[i])
        time.sleep(3)
        elements=browser.find_elements(By.XPATH,"//div[@class='text']")
        elements=elements[0].find_elements_by_class_name('meta')
        dict={}
        dict['title']=browser.find_elements(By.XPATH,'//h2[@class="job-title"]')[0].text
        try:
            dict['company']=browser.find_elements(By.XPATH,"//div[@class='span9 content']//b//a")[0].text
        except:
            continue
        dict['categorie']=browser.find_elements(By.XPATH,"//div[@class='span9 content']")[0].text.split('\n')[1][9:]
        dict['lien']=unquote(jobs[i])
        for j in elements:
            try:
                key=j.find_element_by_tag_name('b').text[:-1]
                if(not key in keys):
                    keys.append(key)
                    print(keys)
                value=j.text.split('\n')[1].replace(',',';')
                dict[key]=value
            except:
                pass
        elementsall.append(dict)
        print(elementsall)
        print('\n')

```

## jobi scraper

```python
import pandas as pd
from urllib.parse import unquote
import re
from selenium import webdriver
from selenium.webdriver.common import keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import csv
browser=webdriver.Chrome("C:\selinium\chromedriver.exe")
wait = WebDriverWait(browser, 10)
elementsall=[]
keys=[]
with open('C:\\Users\\TOSHIBA\\Desktop\\pfe\\jobi.csv','w',encoding="utf-16") as f_object:
    csv_columns =['title','lien','company','categorie','Type de poste','Lieu de travail','Expérience','Étude','Rémunération proposée']
    writer_object = csv.DictWriter(f_object, delimiter=',', fieldnames=csv_columns)
    writer_object.writeheader()
    for page in range (1,40):
        browser.get("https://www.jobi.tn/#!/?q=page="+str(page)+"&sortBy=updated_at&sortOrder=desc")
        time.sleep(10)
        jobs1=browser.find_elements(By.XPATH,'//div[@class="font-1-2"]//a')
        jobs=[]
        time.sleep(5)
        for job in jobs1:
            jobs.append(job.get_attribute('href'))
        print(len(jobs))
        for i in range(len(jobs)):
            print(i+1)
            browser.get(jobs[i])
            time.sleep(5)
            dict={}
            try : 
                dict['title']=browser.find_elements(By.XPATH,'//div//h4')[0].text
            except:
                pass
            try :
                dict['lien']=unquote(jobs[i])
            except:
                pass
            try :
                dict['company']=browser.find_elements(By.XPATH,"//div//a//span[@class='font-light font-1-5 text-white font-uppercase']")[0].text
            except Exception as err:
                print(err)
            try :
                dict['categorie']=browser.find_elements(By.XPATH,"(//div//p//span)[1]")[0].text
            except Exception as err:
                print(err)
            try :
                dict['Type de poste']=browser.find_elements(By.XPATH,"(//div//p//span)[2]")[0].text
            except:
                pass
            try :
                dict['Lieu de travail']=browser.find_elements(By.XPATH,"//div[@class='col m11']")[0].text
            except:
                pass
            try :
                dict['Expérience']=browser.find_elements(By.XPATH,"(//div//span[@class='col m8'])[2]")[0].text
            except:
                pass
            try :
                dict['Étude']=browser.find_elements(By.XPATH,"(//div//span[@class='col m8'])[1]")[0].text
            except:
                pass
            try :
                dict['Rémunération proposée']=browser.find_elements(By.XPATH,"(//div//span[@class='col m8 fx-row'])")[0].text.replace('\n','')
            except:
                pass
            for column in csv_columns:
                if not column in dict:
                    dict[column]='UNKNOWN'
            elementsall.append(dict)
            
            writer_object.writerow(dict)
```
